```python
def Get_communities(data):
  temp={}
  comp=[]
  for key in data:
    for val in data[key]:
      if val in temp:
        temp[val].append(key)
      else:
        temp[val]=[key]
  comp=[tuple(i) for i in temp.values()]
  return comp
```
